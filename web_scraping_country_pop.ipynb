# Web_scrapping with country population of the world 



## Get the data

- For getting the data I use the wikipedia page and scrapping the data from the following link:

https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"
req = requests.get(url).text
soup=BeautifulSoup(req,"html.parser")
table = soup.find_all(class_ = 'wikitable')[0]
```

## Preprocess the data

```python
li = []
for i in soup.find('tbody').select('tr'):
    try:
        text = i.text.split('\n')
        text.remove('')
        li.append(text)
    except:
        pass
```

```python
for i in li:
    i.remove('')
```

```python
links = []
for i in table.find_all("img"):
    links.append(i["src"])    
    
links.insert(169," ")
```

## Making a final Dataframe

```python
df = pd.DataFrame(li[:-1])
df.dropna(axis=1,inplace=True)
df.drop(0,axis=0,inplace=True)
df.columns = ["country", "Region","Subregion","Population(July 2018)",
              "Population(July 2019)","Change"]
```

```python
df
```

## Parsing flag links to dataframe

```python
df["links"]= links
```

## Final Data Frame

```python
from IPython.core.display import HTML
path = df['links']
links = ['<img src="'+ str(x) + '" width="50" >' for x in path]
df['links'] = links
HTML(df.to_html(escape=False))
```
